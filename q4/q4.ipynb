```python
import csv
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

f = open('Seoul_Subway.csv','r',encoding='cp949')
data = csv.reader(f)
next(data)
next(data)

morning_boarding = {}
morning_alighting = {}
morning_total = {}

for row in data:
    for i in range(10, 14):
        row[i] = int(row[i])
    station = row[1] + " " + row[3]
    morning_boarding[station] = morning_boarding.get(station, 0) + (row[10] + row[12])
    morning_alighting[station] = morning_alighting.get(station, 0) + (row[11] + row[13])
    morning_total[station] = morning_total.get(station, 0) + (row[10] + row[11] + row[12] + row[13])

morning_boarding = sorted(morning_boarding.items(), key=lambda x: x[1], reverse=True)[:30]
morning_alighting = sorted(morning_alighting.items(), key=lambda x: x[1], reverse=True)[:30]
morning_total = sorted(morning_total.items(), key=lambda x: x[1], reverse=True)[:30]

morning_boarding_names, morning_boarding_counts = zip(*morning_boarding)
morning_alighting_names, morning_alighting_counts = zip(*morning_alighting)
morning_total_names, morning_total_counts = zip(*morning_total)

plt.bar(morning_boarding_names, morning_boarding_counts, color='blue')
plt.title('출근 시간에 승차한 인원이 가장 많은 지하철역 상위 30개')
plt.xticks(rotation=90)
plt.show()

plt.bar(morning_alighting_names, morning_alighting_counts, color='red')
plt.title('출근 시간에 하차한 인원이 가장 많은 지하철역 상위 30개')
plt.xticks(rotation=90)
plt.show()

plt.bar(morning_total_names, morning_total_counts, color='pink')
plt.title('출근 시간에 승하차한 인원이 가장 많은 지하철역 상위 30개')
plt.xticks(rotation=90)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

def main():
    on = []
    off = []
    onoff = []
    onstation = []
    oncount = []
    offstation = []
    offcount = []
    onoffstation = []
    onoffcount = []
    
    f = open('Seoul_Subway.csv', 'r')
    data = csv.reader(f)
    header = next(data)
    for row in data:
        row = list(map(lambda s: ''.join(s.split()), row))

        a = int(row[9]) + int(row[11])
        b = int(row[10]) + int(row[12])
        c = a + b

        a1 = [row[1] + " " + row[2], a]
        a2 = [row[1] + " " + row[2], b]
        a3 = [row[1] + " " + row[2], c]
        on.append(a1)
        off.append(a2)
        onoff.append(a3)

    on.sort(key=lambda x: -x[1])
    off.sort(key=lambda x: -x[1])
    onoff.sort(key=lambda x: -x[1])

    for i in range(30):
        onstation.append(on[i][0])
        oncount.append(on[i][1])
        offstation.append(off[i][0])
        offcount.append(off[i][1])
        onoffstation.append(onoff[i][0])
        onoffcount.append(onoff[i][1])

    f.close()

    fig, axs = plt.subplots(3, 1, figsize=(10, 20))

    axs[0].bar(onstation, oncount)
    axs[0].set_title("승차")
    axs[0].tick_params(axis='x', rotation=90)

    axs[1].bar(offstation, offcount)
    axs[1].set_title("하차")
    axs[1].tick_params(axis='x', rotation=90)

    axs[2].bar(onoffstation, onoffcount)
    axs[2].set_title("승하차")
    axs[2].tick_params(axis='x', rotation=90)

    plt.show()

if __name__ == "__main__":
    main()
```

```python
import csv
import matplotlib.pyplot as plt

def main():
    file_2022 = open('', encoding='cp949')
    file_2012 = open('2012.csv', encoding='cp949')
    file_2000 = open('2000.csv', encoding='utf-8')

    data_2022 = csv.reader(file_2022)
    data_2012 = csv.reader(file_2012)
    data_2000 = csv.reader(file_2000)

    man = []
    woman = []

    for row in data_2000:
        row[2] = int(row[2])
        row[3] = int(row[3])
        man.append(row[2])
        woman.append(row[3])

    for row in data_2012:
        row[3] = int(row[3])
        row[4] = int(row[4])
        man.append(row[3])
        woman.append(row[4])

    for row in data_2022:
        row[3] = int(row[3])
        row[4] = int(row[4])
        man.append(row[3])
        woman.append(row[4])

    x_data = [2000, 2012, 2022]
    plt.title('Jeju Gender Ratio')
    plt.xlabel("Year")
    plt.ylabel("Gender Ratio")
    plt.plot(x_data, man, color='b', label="Men")
    plt.plot(x_data, woman, color='r', label="Women")
    plt.xticks(x_data)
    plt.legend()
    plt.show()

    file_2022.close()
    file_2012.close()
    file_2000.close()

if __name__ == "__main__":
    main()
```
